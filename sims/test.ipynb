```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Parameters
n = 1
T = 1
Fs = 1000
f = 5

# Time vector
t = np.linspace(0, T, int(T * Fs), endpoint=False)

# Generate square wave signals
y = signal.square(2 * np.pi * f * t)
y = np.tile(y, (n, 1)).T  # Now y has shape (T*Fs, n)
print(y.shape)
# Create subplots
fig, axes = plt.subplots(n, 1, figsize=(10, 6), sharex=True)
if n == 1:
    axes = [axes]
for i in range(n):
    axes[i].plot(t, y[:, i], label=f"Signal {i+1}")
    axes[i].set_ylabel("Amplitude")
    axes[i].legend()
    axes[i].grid()

axes[-1].set_xlabel("Time [s]")
fig.suptitle("Square Wave Signals")

plt.show()
```

```python
t = np.linspace(0, T, int(Fs * T), endpoint=False)  # Time vector
y = np.zeros_like(t)                                # Initialize signal with zeros

impulse_period = int(Fs / f)                        # Samples between impulses
y[::impulse_period] = 1                             # Set impulses at intervals
y = np.tile(y, (n, 1)).T                            # Now y has shape (T*Fs, n)

# Create subplots
fig, axes = plt.subplots(n, 1, figsize=(10, 6), sharex=True)
if n == 1:
    axes = [axes]
for i in range(n):
    axes[i].plot(t, y[:, i], label=f"Signal {i+1}")
    axes[i].set_ylabel("Amplitude")
    axes[i].legend()
    axes[i].grid()

axes[-1].set_xlabel("Time [s]")
fig.suptitle("Square Wave Signals")

plt.show()
```
